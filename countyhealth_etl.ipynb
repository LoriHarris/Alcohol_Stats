### Transform Ranking and State DataFrames

```python
# Rankings
nrank_cols = ["FIPS", "Rank"]
ranking_df = nrank_df[nrank_cols].copy()

# Clean the data by dropping duplicates and setting the index
ranking_df.drop_duplicates("FIPS", inplace=True)
ranking_df.set_index("FIPS", inplace=True)

ranking_df.head()
```

```python
# Create a state_df dataframe that can be a lookup table
state_cols = ["FIPS", "County"]
state_df= nrank_df[state_cols].copy()

# Clean the data by dropping duplicates and setting the index
state_df.drop_duplicates("FIPS", inplace=True)
state_df.set_index("FIPS", inplace=True)

state_df.head()
```

### Transform county DataFrame

```python
rank_cols = ["FIPS", "% Excessive Drinking", "% Alcohol-Impaired"]
rank_transformed= rank_df[rank_cols].copy()

# Rename the column headers
rank_transformed = rank_transformed.rename(columns={"% Excessive Drinking": "x_drink_pct",
                                                    "% Alcohol-Impaired": "etoh_drive_death_pct"})
# Create year variable
# Update Year
rank_transformed['year'] = 2018

# Clean the data by dropping duplicates and setting the index
rank_transformed.drop_duplicates("FIPS", inplace=True)
rank_transformed.set_index("FIPS", inplace=True)

rank_transformed.head()
```

```python
# Merge ranked data with alcohol-related data columns
merged_table = pd.merge(ranking_df, rank_transformed, on="FIPS", how="left")
merged_table.head()
```

### Create database connection

```python
from config import dbuser, dbpasswd, dburi, dbport, dbname

engine = create_engine(f"mysql://{dbuser}:{dbpasswd}@{dburi}:{dbport}/{dbname}")
```

### Load DataFrames into database

```python
state_df.to_sql(name='state_data', con=engine, if_exists='append', index=True)
merged_table.to_sql(name='rank_data', con=engine, if_exists='append', index=True)
```
